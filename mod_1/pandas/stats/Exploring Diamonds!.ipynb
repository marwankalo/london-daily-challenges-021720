**2) Arrange the table so that the diamonds with the highest carats are listed first.**

```python
df = df.sort_values(by='carat', ascending=False)
df
```

**BONUS 1: Use the seaborn library to make a scatter plot between carat and depth, carat and table, carat and price. Use the hue argument to group by either color or cut**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("darkgrid")
```

```python
plt.figure(figsize=(10,18))
plt.subplots_adjust(hspace=0.5)

x = df['carat']
y = df['depth']
w = df['table']
z = df['price']


plt.subplot(3, 1, 1)
sns.scatterplot(x,y, hue=df['cut'])
plt.xlabel('Carat', fontsize=20)
plt.ylabel('Depth', fontsize=20)
plt.title('Carat vs Depth', fontsize=20)

plt.subplot(3, 1, 2)
sns.scatterplot(x,w, hue=df['color'])
plt.xlabel('Carat', fontsize=20)
plt.ylabel('Table', fontsize=20)
plt.title('Carat vs Table', fontsize=20)

plt.subplot(3, 1, 3)
sns.scatterplot(x,z, hue=df['color'])
plt.xlabel('Carat', fontsize=20)
plt.ylabel('Price', fontsize=20)
plt.title('Carat vs Price', fontsize=20)

plt.show()
```
